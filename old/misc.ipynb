```python
rmax, s, th = 100, 1, 5
phi = lambda x, s, th, rmax: rmax/2 * (1 + erf((x - th) / (np.sqrt(2) * s)))
dphi = lambda x, s, th, rmax: rmax/np.sqrt(2*np.pi) / s * np.exp(-(x - th)**2 / (2 * s**2))
```

```python
def inv_phi(r, s, th, rmax):
    x = 2 * r / rmax - 1
    aux = erfinv(x) * np.sqrt(2) * s + th
    # if hasattr(x, '__iter__'):
    #     aux[x<-1+1e-4] = 0
    #     aux[x>1-1e-4] = 1000
    return aux
```

```python
_ = np.linspace(-5, rmax, num=2000)
plt.plot(_, inv_phi(_, s, th, rmax), c='k')
plt.plot(_, inv_phi(_, s + 2, th, rmax), c='r')
```
